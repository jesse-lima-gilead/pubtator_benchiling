```python
from transformers import AutoTokenizer, AutoModel
import torch

# Define models and their respective tokenizers
models = {
    "BioGPT": "microsoft/biogpt",
    #"BioBERT": "dmis-lab/biobert-v1.1",
    #"MedCPT": "path-to-medcpt"  # Replace with actual model name or path
}



# Function to get embeddings
def get_embeddings(model_name, text):
    # Load tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    
    # Tokenize and convert to input IDs
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    
    # Get model output
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Mean pooling of the last hidden state
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings

```

# Generating Embeddings using Multiple Models

```python
# Generate embeddings 
model_name = "BioGPT"
model_path = "microsoft/biogpt"
print(f"Generating embeddings using {model_name}...")
bio_gpt_embeddings = get_embeddings(model_path, text)
print(f"Embeddings from {model_name}: {bio_gpt_embeddings}")
```

# Retrival

```python
user_query = "lung cancer risk from air pollution"

query_embeddings_bio_gpt = get_embeddings("microsoft/biogpt", user_query)
query_embeddings_bio_bert = get_embeddings("dmis-lab/biobert-v1.1", user_query)
```

```python
# Cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between query and text embeddings
similarity_bio_gpt = cosine_similarity(query_embeddings_bio_gpt, bio_gpt_embeddings)

# Calculate cosine similarity between query and text embeddings
similarity_bio_bert = cosine_similarity(query_embeddings_bio_bert, bio_bert_embeddings)

print(f"Similarity between query and text using BioGPT: {similarity_bio_gpt}")

print(f"Similarity between query and text using BioBERT: {similarity_bio_bert}")
```

```python
# Generate initial similarity scores using BioGPT/BioBERT
similarity_scores = similarity_bio_bert[0][0]
threshold = 0.75  # Define a threshold based on experimentation

true_labels = (similarity_scores >= threshold).astype(int)

print(true_labels)
0
```
